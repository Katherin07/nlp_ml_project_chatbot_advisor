Se definen las siguientes funciones para detectar si sobre el corpus hay documentos ilegibles o con problemas de codificación, evidenciando que no se encontraron documentos con estos inconvenientes:

````python
def detect_encoding(text):
    result = chardet.detect(text.encode())
    return result['encoding']

def check_encoding_problems(text):
    try:
        text.encode('utf-8').decode('utf-8')
        return False
    except UnicodeDecodeError:
        return True

def identify_illegible_documents(df):
    illegible_documents = []
    for index, row in df.iterrows():
        for pattern in row['patterns']:
            if check_encoding_problems(pattern):
                illegible_documents.append(index)
                break
    return illegible_documents

illegible_docs = identify_illegible_documents(df)
if not illegible_docs:
    print("No se encontraron documentos ilegibles o con problemas de codificación.")
else:
    print("Documentos ilegibles o con problemas de codificación:")
    print(illegible_docs)
````

## **2. Limpieza de los Textos**
---

 Para la limpieza de los textos del corpus, se seleccionan las operaciones más comunes en distintos preprocesamientos de información textual, como lo son tokenizacion, normalización de textos, lematización, filtrado de palabras como filtrado de stop words, en este caso la limpieza de los textos se realizara con metodos de la libreria ***spacy***, una librería para procesamiento de lenguaje natural (NLP) que provee una forma de uso simple y flexible para implementar soluciones de NLP de forma inmediata:

````python
import spacy
````

````python
# Cargamos spaCy's English language model
nlp = spacy.load("en_core_web_sm")
````

Adicional para complementar la limpieza utilizaremos la libreria ***unidecode*** para normalizar o simplificar el texto del corpus, asi como tambien el filtrado con expresiones regulares:

````python
!pip install unidecode
````

````python
from unidecode import unidecode
import re
````

Se preprocesan los documentos interactuando directamente con el diccionario de intents guardado en la variable *intents_data* y no con el Dataframe utilizado para realizar el analisis exploratorio de los textos, se realiza el preprocesamiento y se generan listas utiles para posterior utilización en la etapa de modelamiento:

````python
words = []
filtered_words = []
classes = []
documents = []
documents_pre = []

for intent in intents_data['intents']:
    for pattern in intent['patterns']:

        # Tokenizamos cada palabra en los patrones usando Spacy
        doc = nlp(pattern)
        w = [token.text for token in doc]

        # Filtramos stopwords, palabras por longitud, texto en minuscula, normalizado y lematizado
        fw = [unidecode(token.lemma_.lower()) for token in doc if not token.is_stop and len(token) > 3 and len(token) < 24]

        # Eliminamos caracteres especiales
        fw_clean = [re.sub(r"[^a-z ]", " ", norm_text.lower()) for norm_text in fw]

        # Eliminamos espacios repetidos
        fw_space = [re.sub(r"\s+", " ", clean_text.lower()) for clean_text in fw_clean]

        words.extend(w)
        filtered_words.extend(fw_space)

        documents.append((w, intent['tag']))
        documents_pre.append((fw_space, intent['tag']))

        if intent['tag'] not in classes:
            classes.append(intent['tag'])
````

Tambien podriamos realizar la limpieza de los textos directamente con el Dataframe utilizado para realizar el analisis exploratorio de los textos, se define una función para el preprocesamiento y se genera el corpus preprocesado en una lista util para posterior utilización en la etapa de extracción de caracteristicas y modelamiento:

````python
pat = re.compile(r"[^a-z ]")
spaces = re.compile(r"\s{2,}")
def preprocess(text, min_len=1, max_len=23):
    # Normalizamos el texto
    norm_text = unidecode(text).lower()

    # Extraemos tokens
    tokens = norm_text.split()

    # Filtramos palabras por longitud
    filtered_tokens = filter(
            lambda token: len(token) >= min_len and len(token) <= max_len,
            tokens
        )
    filtered_text = " ".join(filtered_tokens)
    # Eliminamos caracteres especiales
    clean_text = re.sub(pat, "", filtered_text)
    # Eliminamos espacios duplicados
    spaces_text = re.sub(spaces, " ", clean_text)
    return spaces_text.strip()
````

Debido a que la columna de patrones esta conformada por listas, se define otra función para convertir a string:

````python
# Definir una función para convertir una lista en una cadena de texto
def lista_a_texto(lista):
    return ', '.join(map(str, lista))
````

## **2. Implementación del Embedding**
---


Para la implementación del Embedding primero lo debemos importar desde **sklearn** a partir del módulo `sklearn.feature_extraction.text`, primero `CountVectorizer`:

````python
from sklearn.feature_extraction.text import CountVectorizer
````

Definimos una función para entrenar el vectorizador definiendo parametros de numero maximo de caracteristicas en 150 y la proporción máxima de documentos que contengan un término en 0.7:

````python
def vectorizer_cv(corpus_prep):

  vect_cv = (
    CountVectorizer(max_features=150, max_df=0.7)
    .fit(corpus_prep)
    )

  features_cv = vect_cv.transform(corpus_prep)

  return features_cv, vect_cv
````

````python
features_cv, vect_cv = vectorizer_cv(corpus_prep)
display(features_cv.shape)
````

Extraemos el vocabulario calculado:

````python
vocab_cv = vect_cv.get_feature_names_out()
display(vocab_cv)
````

````python
df_cv = pd.DataFrame(columns=vocab_cv, data=features_cv.toarray())
display(df_cv)
````

Ahora implementamos _TF-IDF_ usando el `TfidfVectorizer` de `sklearn`:

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

Definimos una función para entrenar el vectorizador definiendo parametros de numero maximo de caracteristicas en 150, la proporción máxima de documentos que contengan un término en 0.7, utilizando sublinear scaling:

````python
def vectorizer_tf(corpus_prep):

    vect_tf = TfidfVectorizer(
    max_features=150, max_df=0.7, norm="l2",
    sublinear_tf=True
    ).fit(corpus_prep)

    features_tf = vect_tf.transform(corpus_prep).toarray()

    return features_tf, vect_tf
````

````python
features_tf, vect_tf = vectorizer_tf(corpus_prep)
display(features_tf.shape)
````

````python
df_tf = pd.DataFrame(columns=vocab_tf, data=features_tf)
display(df_tf)
````

Este clasificador es especialmente útil en tareas de clasificación de texto y se utiliza comúnmente en el procesamiento de lenguaje natural (NLP) para tareas como la clasificación de documentos, detección de spam, análisis de sentimientos y más.

---
Las máquinas de vectores de soporte son un modelo de aprendizaje supervisado en el cual se busca representar a los ejemplos en un nuevo espacio, de tal forma que a aquellos ejemplos de diferentes categorías sea posible, en principio, separarlos linealmente en este nuevo espacio.

SVM usa una función conocida como *Kernel*. Intuitivamente, esta función $k$ define qué tan parecidas son dos instancias del conjunto de datos. Formalmente, la función $k$ calcula el producto punto en el espacio de características donde se representarán los datos. Dependiendo del *kernel*, este espacio de características es de mayor dimensionalidad, y facilita la definición de un "*hiperplano*" que separe los ejemplos de ambas características. La dimensión de este *hiperplano* varía de acuerdo al número de características. Si se tienen 2, el *hiperplano* es una recta. Si se tienen 3 el *hiperplano* es un plano en un espacio de 3 dimensiones.

#### **2.2.1 Importar Modelo**
---

Se importa el modelo:

````python
from sklearn.svm import SVC
````

#### **2.2.2 Entrenamiento del Modelo y Selección de Hiperparámetros**

````python
param_grid_svc = {
    'C': [0.1, 1, 10],  # Parámetro de regularización C
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Kernels a explorar
    'degree': [2, 3],  # Grado del kernel polinómico (si se usa)
    'gamma': ['scale', 'auto', 0.1, 1],  # Coeficiente gamma (para kernels 'poly', 'rbf', 'sigmoid')
}
````

Primero realizamos el entrenamiento y la exploración para *CountVectorizer*:

````python
# Crear un objeto GridSearchCV para buscar los mejores hiperparámetros
gsearch_cv_svc = GridSearchCV(
    SVC(),
    param_grid=param_grid_svc,
    cv=5,
    ).fit(features_train_cv, labels_train_cv)
````
